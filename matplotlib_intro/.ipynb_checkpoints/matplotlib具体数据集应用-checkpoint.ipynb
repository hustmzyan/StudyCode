# Matplotlib数据可视化画图

两份数据：
* 美国女性在各个专业的学士学位获取比例的情况
* 鸢尾花数据

## 1.1 引入pandas库后先导入数据：

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('percent-bachelors-degrees-women-usa.csv', index_col = 'Year')
df.info()
```

```python
df.head()
```

# 1.2 绘图

```python
df_CS = df['Computer Science']
plt.plot(df_CS)
```

# 1.3 添加标题和标签

```python
df_CS = df['Computer Science']
plt.plot(df_CS)
plt.title('Percentage of Computer Science\'s degrees awarded to women in the USA')
plt.xlabel('Years')
plt.ylabel('Percentage')
```

# 1.4 绘制多个图形进行对比

```python
df_CS = df['Computer Science']
df_MS = df['Math and Statistics']

df_CS.plot(color='b', style='.-', legend=True)
df_MS.plot(color='r', style='-', legend=True)
plt.title('Percentage of Computer Science\'s degrees VS Math and Statistics\'s ')
plt.xlabel('Years')
plt.ylabel('Percentage')
```

```python
df.plot(alpha = 0.7)
plt.title('Percentage of bachelor\'s degrees to women in the USA')
plt.xlabel('Years')
plt.ylabel('Percentage')
plt.legend(loc='center', bbox_to_anchor=(1.8, 0.5), ncol=2)
plt.axis((1970, 2000, 0, 120))
```

# 2. 散点图

```python
iris = pd.read_csv('iris.csv')
iris.columns = ['seqal_length', 'seqal_width', 'petal_length', 'petal_width', 'species']

iris.plot(kind='scatter', x='seqal_length', y='seqal_width')
plt.xlabel('seqal length in cm')
plt.xlabel('seqal width in cm')
plt.title('iris data analysis')
```

# 3. box箱图

```python
iris.plot(kind='box', y='seqal_length')
plt.ylabel('seqal length in cm')
```

# 4. 柱状图

```python
mask = (iris.species == 'Iris-setosa')
setosa = iris[mask]

setosa.plot(kind = 'hist', bins=50, range=(0,8), alpha=0.5)
plt.title('setosa in iris')
plt.xlabel('CM')
```
